```python
from control.matlab import *
from matplotlib import pyplot as plt
import numpy as np
```

```python
G = tf([1.], [2., 5., 4])
print(G)
```

```python
Kp_init = np.random.rand()
Ki_init = np.random.rand()
Kd_init = np.random.rand()
numd = [Kd_init, Kp_init, Ki_init]
dend = [1., 0]
H = tf(numd, dend)
print(H)
```

```python
target = 1.
sys = feedback(H*G, target)
print(sys)
```

```python
t = np.linspace(0, 50, 3000)
y, T = step(sys, t)
plt.plot(T, y)
plt.grid()
plt.show()
```

```python
from skopt import gp_minimize
from skopt.plots import partial_dependence, plot_convergence, plot_evaluations, plot_objective
```

```python
def loss(G, K, target, t):
    ## PID controler
    numd = K
    dend = [1., 0]
    H = tf(numd, dend)
    
    ## closed system
    sys = feedback(H*G, target)
    
    ## simulation
    y, _ = step(sys, t)
    
    ## error
    e = target-y
    
#     ## inputs into plant
#     u = K[1]*e + np.hstack([K[0]*np.diff(e), 0]) + K[2]*np.cumsum(e)
#     ## error
#     e_value = np.diff(e**2)**2
    ## loss function
    loss = (e**2).sum() + 0.5*(np.array(K)**2).sum()

#            + 5*len(e_value[e_value>1e-5]) ## 収束時間
#            + 1e-3*(u**2).sum() ## 入力の大きさ
    return loss

def objective(params):
    return loss(G, params, 1., t)

space  = [(0., 15.), 
          (0., 15.), 
          (0., 15.)]
K = [Kd_init, Kp_init, Ki_init]

res = gp_minimize(objective, space, x0=K, 
                  n_calls=100, verbose=True,
                  )
print(res.fun)
print(res.x)
```

```python
plot_objective(res)
```

```python
H_res = tf(res.x, [1., 0])
sys_res = feedback(H_res*G, target)
print(sys_res)
y_res, T_res = step(sys_res, t)
print(np.sum((target-y_res)**2))
plt.plot(T_res, y_res)
plt.grid()
plt.show()
```
